### using beautiful soup for importing HTML code

```python
wiki = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
import urllib.request 
response = urllib.request.urlopen(wiki) 
html = response.read()
soup = BeautifulSoup(html,"html.parser")
```

### Create list of of string from HTML

```python
my_table = soup.find('table',{'class':'wikitable sortable'})
#print(my_table)
postalcode = []
borough = []
neighborhood = []
for row in my_table.find_all('tr')[1:]:
    postalcode.append(row.find_all()[0].text)
    borough.append(row.find_all()[1].text)
    neighborhood.append(row.find_all()[2].text)
for a in range(len(postalcode)):
    postalcode[a]=postalcode[a].strip('\n')
    borough[a]=borough[a].strip('\n')
    neighborhood[a]=neighborhood[a].strip('\n')
```

### DataFrame Generation

```python
df=pd.DataFrame(list(zip(postalcode,borough,neighborhood)))
df.columns=['PostalCode','Borough','Neighborhood']
df.drop(df[df['Borough']=='Not assigned'].index, inplace=True)
#print(df)
print(df.shape)
```

### One hot econding and groouping by in order to kluster

```python
# one hot encoding
toronto_onehot = pd.get_dummies(toronto_venues[['Venue Category']], prefix="", prefix_sep="")

# add neighborhood column back to dataframe
toronto_onehot['Neighborhood'] = toronto_venues['Neighborhood'] 

# move neighborhood column to the first column
fixed_columns = [toronto_onehot.columns[-1]] + list(toronto_onehot.columns[:-1])
toronto_onehot = toronto_onehot[fixed_columns]

toronto_onehot.head()# one hot encoding
toronto_onehot = pd.get_dummies(toronto_venues[['Venue Category']], prefix="", prefix_sep="")

# add neighborhood column back to dataframe
toronto_onehot['Neighborhood'] = toronto_venues['Neighborhood'] 

# move neighborhood column to the first column
fixed_columns = [toronto_onehot.columns[-1]] + list(toronto_onehot.columns[:-1])
toronto_onehot = toronto_onehot[fixed_columns]

toronto_onehot.head()

toronto_grouped = toronto_onehot.groupby('Neighborhood').mean().reset_index()
toronto_grouped
```

### Clusterig

```python
# set number of clusters
kclusters = 10

toronto_grouped_clustering = toronto_grouped.drop('Neighborhood', 1)

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(toronto_grouped_clustering)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:10] 
```

### Display the cluster map

```python
address = 'Toronto, ON'

geolocator = Nominatim(user_agent="ny_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Toronto are {}, {}.'.format(latitude, longitude))
```

### Adding labels to the map.

```python
toronto_grouped.insert(0, 'Cluster Labels', kmeans.labels_)
toronto_grouped
toronto_venues_2=toronto_venues
toronto_grouped_2=toronto_grouped[['Cluster Labels','Neighborhood']]
toronto_venues_2 = toronto_venues_2.join(toronto_grouped_2.set_index('Neighborhood'), on='Neighborhood')
toronto_venues_2.head()
```

```python
# create map
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=11)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(toronto_venues_2['Venue Latitude'], toronto_venues_2['Venue Longitude'], toronto_venues_2['Neighborhood'], toronto_venues_2['Cluster Labels']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
